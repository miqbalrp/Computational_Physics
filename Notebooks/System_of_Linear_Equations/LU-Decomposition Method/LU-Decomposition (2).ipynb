## Define Decomposition Function

```python
def decomposition(A):
  # get number of rows from A
  n = len(A)

  # create zeros matrix of L and U
  L = [[0 for row in range(n)]
       for col in range(n)]
  U = [[0 for row in range(n)]
       for col in range(n)]
  
  for p in range(n):
    # upper matrix
    for j in range(p,n):
      # summation of L(p,k)*U(k,j)
      sum = 0
      for k in range(p):
        sum = sum + L[p][k]*U[k][j]
      U[p][j] = A[p][j] - sum

    # lower matrix
    q = p
    for i in range (q,n):
      if (i==q):
        # diagonal of L
        L[i][q]=1
      else:
        # summation of L(i,k)*U(k,q)
        sum = 0
        for k in range(q):
          sum = sum + L[i][k]*U[k][q]
        L[i][q] = (A[i][q] - sum)/U[q][q]

  return L, U
```

## Define Forward Substitution Function

```python
def forward_subs(b,L):
  n = len(b)
  t = [[0 for row in range(n)]
      for col in range(n)]

  for i in range(n):
    sum = 0
    for j in range(i):
      sum = sum + L[i][j]*t[j] 
    t[i] = b[i] - sum
    
  return t
```

## Define Backward Substitution Function

```python
# backward substitution
def backward_subs(t,U):
  n = len(t)
  x = [[0 for row in range(n)]
      for col in range(n)]

  for i in range(n,0,-1):
    sum = 0
    for j in range(i,n):
      sum = sum + U[i-1][j]*x[j]  
    x[i-1] = (1/U[i-1][i-1])*(t[i-1]-sum)

  return x
```

## Define Function to Solve The Linear System

```python
def solve_LU(A,b):
  # factorize matrix A
  L,U = decomposition(A)

  # use vector b and matrix L for forward substitution
  t = forward_subs(b,L)

  # use result from forward substitution and matrix U for backward substitution
  x = backward_subs(t,U)

  return x
```

## Implementing Function to Some Examples



```python
# matrix A and vector b
A = [[3, 2, -1],
     [2, -2, 4],
     [-1, 0.5, -1]]

b = [1,-2,0]

# solve the system using function
x = solve_LU(A,b)

# print the solution
for i in range(len(x)):
  print('x_%d : %.2f' %(i+1,x[i]) )
```

```python
# another example
A = [[1,1,1],
     [4,3,-1],
     [3,5,3]]

b = [1,6,4]

x = solve_LU(A,b)
for i in range(len(x)):
  print('x_%d : %.2f' %(i+1,x[i]) )
```
